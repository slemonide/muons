```python
import matplotlib.pyplot as plt
```

```python
H = [28, 48.16, 25.44, 50.24, 50.23, 37, 21.8, 46, 59.73, 55, 59.26]
Hu = [1, 1, 1, 2, 2, 5, 1, 2, 4, 2, 3]
A = [2, 2.6, 2.6, 2.56, 2.5, 2.57, 2.2, 3.2, 3.7, 3.7, 3.73]
Au = [0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5]
L = [9.65, 13, 9, 9.75, 9.5, 9.5, 7.57, 9.5, 10, 10.5, 10]
Lu = [0.5, 0.5, 0.5, 1, 0.5, 0.5, 1, 0.5, 0.5, 0.5, 0.5]
P = [7, 0, 0, 0, 0, 2.5, 2.25, 2.5, 2.5, 3.5, 2.75]
Pu = [0.5, 0, 0, 0, 0, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5]

labels=['1.3', '1.2', '4.2', '4.1', '4.3', '1.1', '2.1', '3.2', '2.2', '2.3', '3.3']

```

```python
len(Pu)
```

```python
fig, ax = plt.subplots()

ax.errorbar(H, A, xerr=Hu, yerr=Au, fmt='o')
ax.set_xlabel('Depth (mV)')
ax.set_ylabel('Response (mV)')

for label, x, y in zip(labels, H, A):
    ax.annotate(
        label,
        xy=(x, y), xytext=(50, 30),
        textcoords='offset points', ha='right', va='bottom',
        bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
        arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0'))
plt.savefig('HA.png')
```
